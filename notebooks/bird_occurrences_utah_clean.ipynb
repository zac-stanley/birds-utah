```python
# There are way too many records for this dataset to be useful for us. I summarized the data in QGIS and took the average number of counted species
# Plus the five species with occurrences immediately greater and less than the average
# I need to select out that subset now
ebird_subset = ebird_df[(ebird_df['vernacularName'] == 'Wilson\'s Warbler') | (ebird_df['vernacularName'] == 'Bewick\'s Wren')
                        | (ebird_df['vernacularName'] == 'Vesper Sparrow') | (ebird_df['vernacularName'] == 'Loggerhead Shrike')
                        | (ebird_df['vernacularName'] == 'Canyon Wren') | (ebird_df['vernacularName'] == 'Northern Pintail')
                        | (ebird_df['vernacularName'] == 'Clark\'s Nutcracker') | (ebird_df['vernacularName'] == 'Savannah Sparrow')
                        | (ebird_df['vernacularName'] == 'Great Horned Owl') | (ebird_df['vernacularName'] == 'Juniper Titmouse')
                        | (ebird_df['vernacularName'] == 'MacGillivray\'s Warbler')
                       ]
```

```python
ebird_subset.info() # 27,975 records is much more reasonable
```

```python
ebird_subset.to_csv('./data/11birdspecies_southwest.csv')
```
